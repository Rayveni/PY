```python
def TrySearch(pattern,text,default=''):
    try:
        res=pattern.search(text).group(1)
    except:
        res=default
    return res
def Trysplit(pattern,text,default=''):
    try:
        res=pattern.split(text)[0]
    except:
        res=default
    return res
```

```python
pattern_sex = re.compile(r' \((.*?)\) \d')
pattern_leader = re.compile(r' (ed.) ')
pattern_vol=re.compile( r' (\d*?)ml')
df['sex']=df['name'].apply(lambda s:TrySearch(pattern_sex,s,'u'))
df['volume']=df['name'].apply(lambda s:TrySearch(pattern_vol,s,'err'))
df['leader']=df['name'].apply(lambda s:Trysplit(pattern_vol,s,''))
#pattern.split(df.name[8])
#df.name[8],pattern_sex.search(df.name[8]).group(1)
#pattern.search(' (ssjsj) ').group(0)
```
